```python
#using regular expression to remove the repetative word in the ccolumn location of a sinle row
import re
def remove_duplicate_words(string):
    words = re.findall(r'\w+', string)
    words1=[]
    for word in words:
      if word not in words1:
        words1.append(word)
    return ' '.join(words1)
df['location'] = df['location'].apply(remove_duplicate_words)
```
