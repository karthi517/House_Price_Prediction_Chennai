```python
location=[]
bedrooms=[]
bathrooms=[]
area=[]
carpet_area=[]
super_area=[]
ownership=[]
sale_type=[]
age_of_the_property=[]
plot_type=[]
floor=[]
price=[]
furnishing=[]
status=[]
facing=[]
count=[]
```

```python
import time
for i in range(4354):
    time.sleep(3)    
    try:
        x=driver.find_elements(By.TAG_NAME,'h3')[i].click()
    except:
        element = driver.find_element(By.ID, "load_more_button")
        element.click()
        time.sleep(6) 
        x=driver.find_elements(By.TAG_NAME,'h3')[i].click()
    
        
    driver.switch_to.window(driver.window_handles[-1])
    y=driver.find_elements(By.CLASS_NAME,'pdoi-list')
    y0=y[0].text
    z=driver.find_elements(By.CLASS_NAME,'pd-price')
    price.append(z[0].text)
    y00=y0.split('\n')
    try:
        v=driver.find_elements(By.CLASS_NAME,'pd-ap-list')
        v0=v[0].text
        v00=v0.split('\n')
    except:
        v00=[]
    
    if('Location' in y00):
        location.append(y00[y00.index('Location')+1])
    elif('Location :' in v00):
        location.append(v00[v00.index('Location :')+1])
    else:
        location.append(None)
        
    if('Bedrooms' in y00):
        bedrooms.append(y00[y00.index('Bedrooms')+1])
    elif('Bedrooms :' in v00):
        bedrooms.append(v00[v00.index('Bedrooms :')+1])
    else:
        bedrooms.append(None)
        
    if('Bathrooms' in y00):
        bathrooms.append(y00[y00.index('Bathrooms')+1])
    elif('Bathrooms :' in v00):
        bathrooms.append(v00[v00.index('Bathrooms :')+1])
    else:
        bathrooms.append(None)
    
    if('Built Up Area' in y00):
        area.append(y00[y00.index('Built Up Area')+1])
    elif('Built Up Area :' in v00):
        area.append(v00[v00.index('Built Up Area :')+1])
    else:
        area.append(None)
                        
    if('Carpet Area' in y00):
        carpet_area.append(y00[y00.index('Carpet Area')+1])
    elif('Carpet Area :' in v00):
        carpet_area.append(v00[v00.index('Carpet Area :')+1])
    else:
        carpet_area.append(None)
                               
    if('Super Area' in y00):
        super_area.append(y00[y00.index('Super Area')+1])
    elif('Super Area :' in v00):
        super_area.append(v00[v00.index('Super Area :')+1])
    else:
        super_area.append(None)
                              
    if('Ownership' in y00):
        ownership.append(y00[y00.index('Ownership')+1])
    elif('Ownership :' in v00):
        ownership.append(v00[v00.index('Ownership :')+1])
    else:
        ownership.append(None)
    
    if('Sale Type' in y00):
        sale_type.append(y00[y00.index('Sale Type')+1])
    elif('Sale Type :' in v00):
        sale_type.append(v00[v00.index('Sale Type :')+1])
    else:
        sale_type.append(None)   
    
    if('Age of Property' in y00):
        age_of_the_property.append(y00[y00.index('Age of Property')+1])
    elif('Age of Property :' in v00):
        age_of_the_property.append(v00[v00.index('Age of Property :')+1])
    else:
        age_of_the_property.append(None) 
        
    if('Type' in y00):
        plot_type.append(y00[y00.index('Type')+1])
    elif('Type :' in v00):
        plot_type.append(v00[v00.index('Type :')+1])
    else:
        plot_type.append(None) 
    
    if('Property on Floor' in y00):
        floor.append(y00[y00.index('Property on Floor')+1])
    elif('Property on Floor :' in v00):
        floor.append(v00[v00.index('Property on Floor :')+1])
    else:
        floor.append(None) 
    
    if('Furnishing' in y00):
        furnishing.append(y00[y00.index('Furnishing')+1])
    elif('Furnishing :' in v00):
        furnishing.append(v00[v00.index('Furnishing :')+1])
    else:
        furnishing.append(None)
        
    if('Status' in y00):
        status.append(y00[y00.index('Status')+1])
    elif('Status :' in v00):
        status.append(v00[v00.index('Status :')+1])
    else:
        status.append(None)
        
    if('Facing' in y00):
        facing.append(y00[y00.index('Facing')+1])
    elif('Facing :' in v00):
        facing.append(v00[v00.index('Facing :')+1])
    else:
        facing.append(None)    
    time.sleep(2)    
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    
    print(i)
    
```
